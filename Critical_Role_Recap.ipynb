```python
import requests
from bs4 import BeautifulSoup

# Base URL for the recaps
base_url = "https://critrole.com/tag/critical-recap/"

# Headers for scraper
headers = {'User-Agent': 'MyCustomUserAgent/1.0'}
```

```python
# Function to get the HTML content of a page
def get_page_content(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to retrieve page: {url}")
        print(f"Error Code: {response.status_code}")
        return None
    
get_page_content(base_url)
```

```python
# Function to get all recap links from the main page
def get_all_recap_links(base_url):
    html_content = get_page_content(base_url)
    if not html_content:
        return []
    soup = BeautifulSoup(html_content, 'html.parser')
    links = soup.find_all('a', class_='qt-text-shadow')  # Adjust the class name based on actual HTML structure
    recap_links = [link['href'] for link in links]
    return recap_links

recap_links = get_all_recap_links(base_url)

print(recap_links)
```

```python
# Function to add unique recap links to a list
def store_recap_links():
    stringval = "c3e"
    recap_link_bridge = []
    for url in recap_links:
        if stringval in url:
            recap_link_bridge.append(url)
    for url in recap_link_bridge:
        if url not in all_recaps:
            all_recaps.append(url)
    return recap_links, recap_link_bridge, all_recaps
all_recaps = []

store_recap_links()

print(recap_links)
print(all_recaps)
```

```python
# Function to move to the next page
def get_next_page(soup):
    next_button = soup.find('a', text='Next')
    if next_button:
        return next_button['href']
    return None
```

```python
# Function to find on the first episode
def extract_substring(input_string, start_delim, end_delim):
    # Split the string at the start delimiter
    parts = input_string.split(start_delim)
    
    # Check if the start delimiter is found
    if len(parts) > 1:
        # Split the remaining part at the end delimiter
        sub_parts = parts[1].split(end_delim)
        
        # Check if the end delimiter is found
        if len(sub_parts) > 1:
            return sub_parts[0]
    
    # Return None if delimiters are not found
    return None

def first_episode():
    # Find the latest or last episode of the campaign
    start_delim = "c3e"
    end_delim = "-"
    ep_count = extract_substring(all_recaps[0], start_delim, end_delim)
    return ep_count

```

```python
# Main function to scrape all recaps
def scrape_all_recaps(base_url):
    recap_links = get_all_recap_links(base_url)
    all_recaps = []
    for link in recap_links:
        html_content = get_page_content(link)
        if html_content:
            recaps = parse_recaps(html_content)
            all_recaps.extend(recaps)
    return all_recaps
```

```python
# Run the scraper
all_recaps = scrape_all_recaps(base_url)
for i, recap in enumerate(all_recaps, 1):
    print(f"Recap {i}:\n{recap}\n")
```
